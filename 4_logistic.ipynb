```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Read the dataset
df = pd.read_csv('logistic_data_50.csv')
X = df[['X']]
y = df['Y']

# Fit the Logistic Regression model
model = LogisticRegression()
model.fit(X, y)

# Predict probabilities for a smooth sigmoid curve
X_test = np.linspace(0, 55, 60).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(X_test, y_prob, color='black', linewidth=2, label='S-shaped curve')
plt.scatter(X, y, color='red', label='Data points')
plt.axhline(y=0, color='orange', linewidth=2)
plt.axhline(y=1, color='skyblue', linewidth=2)
plt.axvline(x=0, color='orange', linewidth=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Logistic Regression from CSV Data')
plt.legend()
plt.grid(True)
plt.show()
```
